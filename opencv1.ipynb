```python
#rescale

def rescaleFrame(frame,scale=0.4):
    width=int(frame.shape[1]*scale) #frame.shape[1] is the width of the original vid or img
    height=int(frame.shape[0]*scale) #frame.shape[0] is the height of the original vid or img

    dimensions=(width,height)    

    return cv2.resize(frame,dimensions,interpolation=cv2.INTER_AREA)

#rescaling videos (only work for videos from web/live camera)

def changeRes(width,height):
    capture.set(3,width)
    capture.set(4,height)
```

```python
#Dilating the img

img=cv2.imread('imgs/sage.jpg')
img=rescaleFrame(img)
cv2.imshow('sage',img)

canny= cv2.Canny(img,125,175)
cv2.imshow('canny',canny)

dilated=cv2.dilate(canny,(3,3),iterations=3)
cv2.imshow('dilated',dilated)





cv2.waitKey()
```

```python
#Eroding an img

eroded=cv2.erode(dilated,(3,3),iterations=3)
cv2.imshow('eroded',eroded)

cv2.waitKey()
```

Contour Detection

```python
#Contour detection

img=cv2.imread('imgs/cats.jpg')
#img = cv2.resize(img,(500,600))
cv2.imshow('cats',img)



gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow('cats_gray',gray)

#ret,thresh =cv2.threshold(gray,125,255,cv2.THRESH_BINARY)
#cv2.imshow('thresh',thresh)


blur=cv2.GaussianBlur(gray,(3,3),cv2.BORDER_DEFAULT)
cv2.imshow('blur',blur)


canny= cv2.Canny(blur,125,175)
cv2.imshow('canny',canny)

contours,hierarchies=cv2.findContours(canny,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)  #CONTOURS=A PYTHON LIST OF ALL THE COORDINATES IF THE CONTOURS THAT WHERE FOUND IN THE IMG ||| HIERARCHIES= THE HIERACHICAL REPRESENTATION OF CONTOURS
print(f"{len(contours)} countour(s) found!")


blank=np.zeros(img.shape,dtype='uint8')
cv2.imshow('blank',blank)



cv2.drawContours(blank,contours,-1,(0,0,200),1) #1 means that we want all the contours and 2 is the thickness
cv2.imshow('contours_draw',blank)











cv2.waitKey()
```

 BITWISE operations

```python
#BITWISE 

blank=np.zeros((400,400),dtype='uint8')
rectangle=cv2.rectangle(blank.copy(),(30,30),(370,370),255,-1)
circle=cv2.circle(blank.copy(),(200,200),200,255,-1)

cv2.imshow('rectangle',rectangle)
cv2.imshow('circle',circle)


#Bitwise AND (intersection)

bitwise_and=cv2.bitwise_and(rectangle,circle)
cv2.imshow('bitwise_and',bitwise_and)


#Bitwise OR  (intersaction+nonintersaction)

bitwise_or=cv2.bitwise_or(rectangle,circle)
cv2.imshow('bitwise_or',bitwise_or)


#Bitwise XOR (non intersaction)

bitwise_xor=cv2.bitwise_xor(rectangle,circle)
cv2.imshow('bitwise_xor',bitwise_xor)


#Bitwise NOT

bitwise_not=cv2.bitwise_not(rectangle)
cv2.imshow('bitwise_not',bitwise_not)


cv2.waitKey()
```

```python
#Face Detection Haar_Cascades (image)



img=cv2.imread('imgs/people1.jpg')
img= cv2.resize(img,(750,500))

gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)


haar_cascade=cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
faces_rect=haar_cascade.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=4)
    
for(x,y,z,h) in faces_rect :
    cv2.rectangle(img,(x,y),(x+z,y+h),(0,255,0),2)


cv2.imshow('Detected faces ',img)

print(f"Number of faces found = {len(faces_rect)}")




cv2.waitKey()
```

```python
#Tracking

from tracker import*

tracker = EuclideanDistTracker() #this tracker takes all the bounding boxes of the objects into oone array

cap =cv2.VideoCapture("videos/highway.mp4")

object_detector=cv2.createBackgroundSubtractorMOG2(history=100,varThreshold=50) #extrate the moving objects from the stable camera


while(True): #tant la camera est allum√©e
    ret,frame=cap.read() #sortie de la fonction = ret and frame , ret = if there is frame i
    #gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY) Q

    #height,width,_ = frame.shape
    #print(height,width)

    # Extract region of interest
  
    region=frame


    #1 object detection
    mask=object_detector.apply(region)
    cv2.imshow('mask',mask)

    threshold,thresh=cv2.threshold(mask,254,255,cv2.THRESH_BINARY)
    
    
    
    contours,hierarchies=cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)


    detections=[]

    for cnt in contours:
        


        #Calculate area and remove all the small elmnts

        area=cv2.contourArea(cnt)
        if area > 100:
            #cv2.drawContours(region,[cnt],-1,(0,255,0),2)
             
            x,y,w,h =cv2.boundingRect(cnt)
            #cv2.rectangle(region,(x,y),(x+w,y+h),(0,255,0),2)
            print(x,y,w,h)

            detections.append([x,y,w,h])
            #print(detections)



    #2 object tracking

    boxes_ids=tracker.update(detections) # boxesid is detection list plux the tracker
    #print( boxes_id)
    for box_id in boxes_ids:
        x,y,w,h,id=box_id
        cv2.putText(region,str(id),(x+w//2,y-15),cv2.FONT_HERSHEY_PLAIN,2,(0,250,0),2)
        cv2.rectangle(region,(x,y),(x+w,y+h),(0,255,0),2)



    cv2.imshow('region',region)
    cv2.imshow('frame',frame)

    if cv2.waitKey(1) &  0xff==ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
